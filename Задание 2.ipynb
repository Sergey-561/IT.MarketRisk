```python
#подключение к базе данных для дальнейших пунктов
import pandas as pd
import pyodbc 
server = 'ACER-PC\SQLEXPRESS' 
database = 'MarketRisk' 
username = 'Sergey-561' 
password = '457457aa' 
cnxn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER='+server+';DATABASE='+database+';UID='+username+';PWD='+ password)
```

```python
Query1 = ("""SELECT * FROM [MarketRisk].[dbo].[bond_information] WHERE [FaceFTName] = 'RUB'""" ) #запрос к базе данных для вывода необходимой информации
```

```python
securities = pd.read_sql(Query1, cnxn)
```

```python
securities
```

```python
#создание таблички облигаций по конкретному эмитенту в зависимости от конкретной даты, по полю date
BPsecurities = securities[(securities['EndMtyDate'] > date) & (securities['BegDistDate'] < date) & (securities['IssuerName'] == issuer)]
```

```python
#вставим в новую табличку для bar plot введённую ранее дату 
BPsecurities.insert(3,'date',date) 
```

```python
#добавим в таблицу столбец с временем, которое осталось до погашения облигаций
BPsecurities['Время до погашения'] = BPsecurities['EndMtyDate'] - BPsecurities['date'] 
```

```python
import numpy as np
BPsecurities['Время до погашения']= (BPsecurities['Время до погашения'] / np.timedelta64(1, 'D')).astype(int)
```

```python
BPsecurities['Время до погашения'] = BPsecurities['Время до погашения']/365
```

```python
BPsecurities
```

```python
import matplotlib.pyplot as plt
Объём = BPsecurities['SumMarketVal']
Days_to_maturity=BPsecurities['Время до погашения']
plt.bar(Days_to_maturity,Объём, width = 0.8, align = 'edge')
plt.xlabel('Лет до погашения')
plt.ylabel('Объём')
plt.title('Структура для '+str(issuer) +' на ' + str(date))
plt.show()
```

```python
#фильтрация облигаций в зависимости от количества лет, которое осталось до погашения
BPsecurities['меньше 5-ти лет'] = BPsecurities['Время до погашения'] < 5
```

```python
BPsecurities['больше 10-ти лет'] = BPsecurities['Время до погашения'] > 10
```

```python
a = BPsecurities.loc[BPsecurities['меньше 5-ти лет']== True, 'SumMarketVal'].sum()
b = BPsecurities.loc[(BPsecurities['меньше 5-ти лет']== False) & ((BPsecurities['больше 10-ти лет']== False)), 'SumMarketVal'].sum()
c = BPsecurities.loc[BPsecurities['больше 10-ти лет']== True, 'SumMarketVal'].sum()
```

```python
Volumes = [a,b,c]
Groups = ['меньше 5-ти лет', 'от 5-ти до 10-ти лет', 'больше 5-ти лет']
```

```python
plt.pie( Volumes, labels= Groups, shadow=False, autopct='%1.1f%%',)
plt.title('Структура погашения для '+str(issuer) +' на ' + str(date))
plt.show()
```

```python
#фильтрация облигации из всего списка
BPsecurities = BPsecurities[BPsecurities['ISIN, RegCode, NRDCode'] == ISIN]
```

```python
BPsecurities [['ISIN, RegCode, NRDCode', 'SecurityType' ,'CouponType', 'HaveOffer']]
```

```python
#построение таблицы по структуре платежей по облигации, с учётом столбца coupon из базы SQL
if BPsecurities['CouponType'].all() == 'Постоянный' and BPsecurities['HaveOffer'].all() == 0 and BPsecurities['AmortisedMty'].all() == 0:
    structure = pd.DataFrame(data={'дата': [], 'платёж': []})
    from datetime import datetime, timedelta
    j = BPsecurities.loc[:,'EndMtyDate'].astype(object).all()
    i=0
    structure.at[i, 'дата'] = j
    structure.at[i, 'платёж'] = BPsecurities['coupon'] + BPsecurities['FaceValue']
    while j >= BPsecurities['date'].astype(object).all():
        j -= timedelta(days=365/BPsecurities.loc[:,'CouponPerYear'].sum())
        i += 1
        structure.at[i, 'дата'] = j
        structure.at[i, 'платёж'] = BPsecurities['coupon']
    print(structure)
else:
    print('Облигация не подходит: тип ее купона не является постоянным, в облигацию встроена аферта или же она является амортизируемой')
```

```python
#построение диаграммы по структуре платежей, если облигация подходит
import matplotlib.pyplot as plt
fig = plt.figure()
plt.bar(structure['дата'].values,structure['платёж'].values, width= 10)
plt.xlabel='Дата платежа'
plt.ylabel='Размер платежа'
plt.title('Структура платежей по облигации с ISIN '+str(ISIN)+' эмитента '+str(issuer) +' на дату ' + str(date))
fig.autofmt_xdate()
plt.show()
```
